```python
 
for i,j in zip(df.ClassId, df.Path):
    if not os.path.exists("Test_all_foldered/{}/".format(i)):
        os.makedirs("Test_all_foldered/{}/".format(i))
    print(i)
    shutil.copy2(j,"Test_all_foldered/{}/".format(i))
```
