```python
import numpy as np
import pandas as pd
from PIL import Image
```

```python
annotations_file = 'annotations_split.csv'
```

```python
anno = pd.read_csv(annotations_file)
anno.head()
```

```python
# Get filenames from annotations file and get size of each image

width_list = []
height_list = []

filename = []
x1 = []
y1 = []
x2 = []
y2 = []
classes = []


for i in anno["filename"]:
    image = Image.open(i)

    width, height = image.size
    
    if (width!=704):
        width_list.append(width)
        height_list.append(height)

# Insert width and height of images to annotations data frame
anno['width'] = width_list
anno['height'] = height_list

anno.head()
```

```python
annos = anno[anno['width'] > 650] 
annos
```

```python
# Save dataframe as csv file
anno.to_csv("annotations_imSize.csv", index=False)
```
